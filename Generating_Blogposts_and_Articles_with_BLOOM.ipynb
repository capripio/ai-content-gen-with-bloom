```python
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
```

```python
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python

model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-3b", use_cache=True)
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-3b")
```

```python
set_seed(424242)
```

```python
 model.__class__.__name__
```

```python
prompt = f'Polyester. Soft and and comfortable to wear.\nA flannel hooded shirt jacket with classic plaid element outer and fuzzy fleece lining inside design adds a relaxed and chic appearance, stylish but also keep you warm. Suitable for women or men, also for couples outerwear.\nThis unisex casual jacket features long sleeves, plaid pattern, flap chest pockets, button closure, a drawstring hood and flannel lining (the sleeves are not lined).'
```

```python
input_ids = tokenizer(prompt, return_tensors="pt").to(0)
```

```python
sample = model.generate(**input_ids, max_length=200, num_beams = 2, num_beam_groups = 2, top_k=1, temperature=0.9, repetition_penalty = 2.0, diversity_penalty = 0.9)
```

```python

print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```
